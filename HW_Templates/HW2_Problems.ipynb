Use zero- through fourth-order Taylor series expansions to predict $f(2.5)$ for $f(x)=ln(x)$ using a base point at $a=1$. Comput the true percent relative error $\varepsilon_t$ for each approximation. Discuss the meaning of the results.

<font color='red'> **TL;DR**:
Explain how the zero- through fourth-order Taylor series approximations can be calculated. Create a program that computes these approximations for the function ln(x) using inputs for the base point (a), evaluation point (x), and maximum order (n), and also calculates the relative error for each approximation. The program should output the Taylor series approximations and percent errors for each order form 0 to n. Print the results for orders 0 through 4. Discuss the results and provide an explanation for the trend in relative errors as a function of Taylor series order.
</font>

### <font color='blue'>**Solution 4.6**</font>

```python
import math
# Define the function ln(x)
def ln(x):
  return math.log(x)

# Define the Taylor series expansion for ln(x) around a specific point a
def taylor_ln(x, a, n):
  result =  0.0
  err = 0.0
  actual = ln(x)
  expansion = []
  error = []
  for k in range(n):
    result += ((-1) ** k) * (x - a) ** (k + 1) / ((k + 1) * a ** (k + 1))
    expansion.append(result)
    err = abs(actual - result) / abs(actual) * 100
    error.append(err)
  return result, expansion, error

# Specify the value of x and the point a around which to expand the series
x = 2.5
a = 1.0
n = 4
# Number of terms in the Taylor series expansion, ie order
actual = ln(x)
zero_order = math.log(a)
zero_error = abs(actual - zero_order) / abs(actual) * 100
taylor_result, expansion, error = taylor_ln(x, a, n)
print("Actual Answer:", actual)
print("Zero order: ",zero_order," error percentage: ",zero_error)
print("Taylor Series Expansion:", expansion)
print("Error Percentages:", error)
```

**Discussion of Results**

[To calculate the zero through fourth order Taylor series approximations, we start off with Taylor series formula which will express a function in terms of its derivatives evaluated at the expansion point.  For ln(x), the derivatives follow a simple pattern of f^(k) (x) = (-1)^k-1 * (k-1)! / x^k, k >= 1.  We are going to evaluate all derivatives at a=1 and we should get f'(1) =0, f'(1) = 1, f"(1) = -1, f^3 (1) = 2, f^4 (1) = -6.  Next we plug these into Taylor series formula.  

The results show that the Taylor series approximations for Ln(x) about a = 1 don't become more accurate as the order increases when evaluated at x = 2.5.  What we notice is that the approximations flunctuate in terms of overestimating and underestimating the true value.  However we do notice that the percent error increases for the third and fourth order approximations.  Since the x= 2.5 lies outside of radius convergence of |x-1| < 1, the Taylor polynomial doesn't approach the true function value.  What this ends up leading to is the relative errors failing to decrease with increasing order.  What this reveals is that the evaluation point has to be within the interval of convergence for the Taylor series to be reliable.]

Interpret your results on the basis of the remainder term of the Taylor series expansion.

<font color='red'> **TL;DR**:
Explain how the forward difference, backward difference, and centered difference methods work. Create a program that executes these methods on the function above and compares the accuracy of these results to the true derivative. ### <font color='blue'>**Solution 4.7**</font>

[link text](https://)

```python
import math

# Given function
def function(x):
    return 25*x**3 - 6*x**2 + 7*x - 88

# True derivative
def derivative(x):
    return 75*x**2 - 12*x + 7

# Parameters
x = 2.0
h = 0.2
actual = derivative(x)

# Finite difference approximations
forward = (function(x + h) - function(x)) / h
backward = (function(x) - function(x - h)) / h
centered = (function(x + h) - function(x - h)) / (2*h)

# Relative percent errors
err_forward = abs((actual - forward)/actual) * 100
err_backward = abs((actual - backward)/actual) * 100
err_centered = abs((actual - centered)/actual) * 100

# Print results
print(f"True derivative f'(2) = {actual:.6f}\n")

print("Method        Approximation     Relative Error (%)")
print("--------------------------------------------------")
print(f"Forward       {forward:.6f}        {err_forward:.6f}")
print(f"Backward      {backward:.6f}        {err_backward:.6f}")
print(f"Centered      {centered:.6f}        {err_centered:.6f}")
```

**Discussion of Results**

[The forward difference method uses point ahead of x and approximates derivative by f'(x) = ((f(x+h) - f(x)) / h).  Backward difference method works similarly but uses point behind x, giving f'(x) = (f(x) - f(x-h) / h), and it also has an error of order O(h).  The centered difference uses points on both sides of x and subtracts Taylor expansions of f(x+h) and f(x-h) to cancel first order terms.  The equation for that is f'(x) = (f(x+h) - f(x-h) / 2h) and is probably the most precise method.
  To compare the accuracy, you compute each numerical derivative at x=2 and then compare it to the true derivative f'(2) using the relative percent error formula.  The method that has the smallest relative percent error will be the most accurate.  The centered difference is the most accurate method for same step size because when calculated with x=2 and h=0.2, it had the lowest error percentage at 0%.]

### <font color='blue'>**Solution 24.19**</font>



```python
import numpy as np

# Distance stations (m) and depths (m) read from Fig. P24.19
x = np.arange(0, 22, 2)  # 0,2,...,20
d = np.array([0, 1.8, 2.0, 4.0, 4.0, 6.0, 4.0, 3.6, 3.4, 2.8, 0.0])

def trap_composite(y, h):
    """Composite trapezoidal rule for equally spaced data."""
    return h * (0.5*y[0] + y[1:-1].sum() + 0.5*y[-1])

def simpson_13(y, h):
    """Composite Simpson's 1/3 rule for equally spaced data (needs even # intervals)."""
    n = len(y) - 1
    if n % 2 != 0:
        raise ValueError("Simpson's 1/3 requires an even number of intervals.")
    odd_sum = y[1:n:2].sum()
    even_sum = y[2:n:2].sum()
    return (h/3) * (y[0] + y[-1] + 4*odd_sum + 2*even_sum)

# Trapezoid h=2 (all data)
A_trap_2 = trap_composite(d, h=2)

# Trapezoid h=4 (every other point)
d4 = d[::2]  # indices 0,2,4,6,8,10 -> x=0,4,8,12,16,20
A_trap_4 = trap_composite(d4, h=4)

# Simpson h=2
A_simp_2 = simpson_13(d, h=2)

print("Area estimates for stream cross-section:")
print(f"Trapezoidal (h=4 m):  {A_trap_4:.2f} m^2")
print(f"Trapezoidal (h=2 m):  {A_trap_2:.2f} m^2")
print(f"Simpson 1/3 (h=2 m):  {A_simp_2:.2f} m^2")
```

**Discussion of Results**

[Trapezoidal Rule approximates the area under a curve by dividing the interval into small segments and replacing the curve on each segment with a line.  The Simpson's Rule replaces each pair of intervals with a parabolic curve rather than a straight line.  For many intervals, it follows formula of: intergal of a to b, f(x)dx = (change in x / 3) *( f(x0 + 4(f1 + f3 +... +fn-1) + 2(f2 +f4+...+fn-2)+f(xn)))

The trapezoidal rule with h = 4m is the least accurate because it uses fewer points and assumes the depth changes linearly over long 4-m spans.  Therefore it ends up missing important curvature in the channel shape.  The trapezoidal rule with h = 2m is more accurate because the smaller captures more of the cross-section detail. Simpson's 1/3 rule with h=2m is typically the most accurate of the three because it fits parabolic quadratic segments over pairs of intervals which represents curvature much better than straight lines.]

### <font color='blue'>**Solution 24.21**</font>



```python
#Importing Needed Packages
%reset -f
import numpy as np
import matplotlib.pyplot as plt

#Given Information
dx = 0.125 #mm        this is step size
L = 3 #m              beam length
E = 200e9 #N/m^2      modulus of elasticity
w0 = 2.5e5 #N/m     initial distribution load
I = 0.0003 #m^4       moment of inertia

#Setting Up
x = np.arange(0,L,dx)
y = np.zeros(len(x))
center_m = np.zeros(len(x))
center_v = np.zeros(len(x))

#Defining the slope Function
def theta(x):
  return (w0/(120*E*I*L)*(-5*x**4+6*L**2*x**2-L**4))


# integrate to find deflection using trapezoid rule
for i in range(len(x)-1):
  h = dx
  y[i+1] = y[i] + (h/2) * (theta(x[i]) + theta(x[i+1]))

#Defining the Moment Function
for i in range(len(x)):
  center_m[i] = E * I * (theta(x[i] + dx) - theta(x[i] - dx)) / (2 * dx)

def moment(x):
  return I*w0/(120 * I * L) * (-20 * x**3 + 12 * L**2 * x)

#Defining the Shear Function
# this could also use the difference between the numerical moment values, center_m
for i in range(len(x)):
  center_v[i] = (moment(x[i] + dx) - moment(x[i] - dx)) / (2 * dx)

  #Plotting the Deflection Graph
plt.figure(figsize=(10,6))
plt.plot(x, y)
plt.xlabel('Distance (m)')
plt.ylabel('Deflection (m)')
plt.title('Beam Deflection')
plt.grid(True)
plt.show()
#Plotting the Moment Graph
plt.figure(figsize=(10,6))
plt.plot(x, center_m)
plt.xlabel('Distance (m)')
plt.ylabel('Moment (N*m)')
plt.title('Moment Diagram')
plt.grid(True)
plt.show()
#Plotting the Shear Graph
plt.figure(figsize=(10,6))
plt.plot(x, center_v)
plt.xlabel('Distance (m)')
plt.ylabel('Shear (N)')
plt.title('Shear Diagram')
plt.grid(True)
plt.show()
```
